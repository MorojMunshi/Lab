```python
from transformers import pipeline
#Basic Text Generation
#We’ll start with a small model (gpt2) just to show the idea.

from transformers import pipeline

generator = pipeline(
    task="text-generation",
    model="gpt2",           # small, runs on CPU
    max_new_tokens=80
)

prompt = "Artificial intelligence in education will"
output = generator(prompt)
print(output[0]["generated_text"])
```

```python
tests = [
    "Translate to Arabic: Data science empowers decision making.",
    "Give me 3 research ideas about using LLMs in higher education.",
    "Summarize: Large language models are powerful because..."
]

for t in tests:
    print("PROMPT:", t)
    print("ANSWER:", generator(t)[0]["generated_text"])
    print("-"*50)
```

```python
#Prompt Engineering Mini-demo
instruction_1 = "Write a short paragraph about blockchain."
instruction_2 = "You are a professor in data science. Write a formal paragraph about blockchain for undergraduate students."
instruction_3 = "Write 3 bullet points explaining why blockchain is useful in quality and accreditation."

for ins in [instruction_1, instruction_2, instruction_3]:
    print("PROMPT:", ins)
    print(generator(ins)[0]["generated_text"])
    print("="*80)
```

```python
#Summarization Task
#We can use FLAN-T5 to summarize a long text.
text = """
Large language models (LLMs) are neural networks trained on vast amounts of text data.
They can generate human-like text, translate languages, answer questions, and assist in coding.
Despite their power, they may hallucinate or produce biased answers if the training data is biased.
"""

summary_prompt = f"Summarize this in 3 bullet points:\n{text}"
print(generator(summary_prompt)[0]["generated_text"])
```

```python
#Question Answering over a Given Context
#Here we show: “LLM answers based on a passage”.
context = """
Umm Al-Qura University is a leading Saudi university that adopts digital transformation and AI in education.
The College of Computing offers programs in data science, AI, and cybersecurity.
The department aims to improve quality through automated tools and chatbots.
"""

question = "What does the college aim to improve?"

qa_prompt = f"Answer the question based on the text.\nText: {context}\nQuestion: {question}"
print(generator(qa_prompt)[0]["generated_text"])
```

```python
#Tiny Chat Loop (console)
from transformers import pipeline

bot = pipeline("text2text-generation", model="google/flan-t5-small")

print("LLM Chatbot (type 'exit' to quit)\n")
while True:
    user = input("You: ")
    if user.lower() in ["exit", "quit"]:
        break
    prompt = f"Answer the following user politely and briefly:\n{user}"
    reply = bot(prompt)[0]["generated_text"]
    print("Bot:", reply)
```
